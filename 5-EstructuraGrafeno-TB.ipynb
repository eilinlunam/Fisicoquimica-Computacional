# Estructura periódica del Grafeno.

$$ \vec{a_1} = \frac{3}{2} acc \hat{i} + \frac{\sqrt{3}}{2} acc  \hat{j}  $$

$$ \vec{a_2} = \frac{3}{2} acc \hat{i} - \frac{\sqrt{3}}{2} acc \hat{j}   $$


$$ \vec{A} = 0 \hat{i} + 0 \hat{j} $$
$$ \vec{B} = acc \hat{i} + 0 \hat{j} $$

Con todo espacio real, tenemos asociado un espacio recíproco:

$$ \vec{k} = n \vec{b_1} + m \vec{b_2} $$

Los vectores $\vec{b_1}$ y $\vec{b_2}$ son las bases del espacio recíproco.

Los vectores del espacio real y el espacio recíproco están relacionados por medio de la siguiente relación cíclica:

$$ \vec{b_1} = \frac{2 \pi (\vec{a_2} \times \vec{a_3} ) }{| \vec{a_1} \cdot ( \vec{a_2} \times \vec{a_3})|} $$

$$ \vec{b_3} = \frac{2 \pi (\vec{a_1} \times \vec{a_2} ) }{| \vec{a_1} \cdot ( \vec{a_2} \times \vec{a_3})|} $$

$$ \vec{b_2} = \frac{2 \pi (\vec{a_3} \times \vec{a_1} ) }{| \vec{a_1} \cdot ( \vec{a_2} \times \vec{a_3})|} $$

## Grafeno en Tight Binding

+ Para calcular las energías del sistema:

$$ \hat{H} | \Psi > = E | \Psi > $$

+ En la aproximación de TB,

$$ \Psi = \sum_{i,l} C_{i,l} (\vec{k}) \chi_{i,l}(\vec{k}, \vec{r}) $$

+ Como estamos en un sistema periódico,

$$ \chi_{i,l} (\vec{k}, \vec{r}) = \frac{1}{\sqrt{N}} \sum_{\vec{R_i}} e^{i \vec{k} \cdot \vec{R_i}} \Phi_l (\vec{r}-\vec{R_i})  $$

+ Para el grafeno, tomaremos los orbitales $P_z(\pi)$

$$ \chi_i(\vec{k}, \vec{r}) = \frac{1}{\sqrt{N}} \sum_{\vec{R_i}} e^{i \vec{k} \cdot \vec{R_i}} \Phi (\vec{r} - \vec{R}_i)$$

$$ \Psi = \sum_i C_i (\vec{k}) e^{i \vec{k} \cdot \vec{R}_i } \Phi (\vec{r}-\vec{R}_i) $$

$$ \hat{H} \sum_i C_i (\vec{k}) e^{i \vec{k} \cdot \vec{R_i}} \Phi (\vec{r} - \vec{R_i}) = E \sum_i C_i (\vec{k}) e^{i \vec{k} \cdot \vec{R}} \Phi (\vec{r} - \vec{R_i}) $$

$$ \sum_i C_i(\vec{k}) e^{i \vec{k} \cdot (\vec{R}_i - \vec{R}_i}) \int  \Phi^{*} (\vec{r}-\vec{R}_j) \hat{H} \Phi(\vec{r}-\vec{R}_i) d\vec{r} = E \sum_i C_i (\vec{k}) e^{i \vec{k} \cdot (\vec{R}_i - \vec{R}_j)} \int \Phi^{*}(\vec{r} - \vec{R}_j) \Phi(\vec{r} - \vec{R}_i) d\vec{r} $$

1) Aproximación a primeros vecinos.  
2) Suponemos que nuestra base es ortogonal.  

$$ \sum_i (i (\vec{k}) e^{i \vec{k} \cdot (\vec{R_i} - \vec{R_j}) } \int \Phi ^{*} (\vec{r} - \vec{R}_j ) \hat {H} \Phi (\vec{r} - \vec{R}_i ) d\vec{r} = E \sum_i C_i(\vec{k}) e^{i \vec{k} \cdot (R_i - R_j) } \delta_{ij}$$

+ $j$ correrá sobre todos los átomos de la celda unitaria j=A,B.


$$ (1) $$

$$ \sum_i C_i (\vec{k}) e^{i \vec{k} \cdot (\vec{R}_i - \vec{R}_A} t_{Ai} = E C_A (\vec{k}) C_B (\vec{k}) e^{i \vec{k} \cdot (\vec{R}_i - \vec{R}_A) } t_{Ai} = E C_A (\vec{k}) $$ 

$$ C_B (\vec{k}) e^{i \vec{k} \cdot acc \hat{i} } t + C_B (\vec{k}) e^{i \vec{k} \cdot \left(-\frac{acc}{2} \hat{i} + \frac{\sqrt{3}}{2} acc \hat{j} \right)} t + C_B (\vec{k}) e^{i \vec{k} \cdot \left(-\frac{acc}{2} \hat {i} - \frac{\sqrt{3}}{2} acc \hat{j} \right)} t + C_A (\vec{k}) E_s = E C_A (\vec{k}) $$

$$ C_B (\vec{k}) \left( e^{i \vec{k} \cdot acc \hat{i} } + e^{i \vec{k} \cdot \left(-\frac{acc}{2} \hat{i} + \frac{\sqrt{3}}{2} acc \hat{j} \right)} + e^{i \vec{k} \cdot \left(-\frac{acc}{2} \hat {i} - \frac{\sqrt{3}}{2} acc \hat{j} \right)} \right) t + C_A (\vec{k}) E_s = E C_A (\vec{k}) $$

$$ (2) $$

$$ C_A (\vec{k}) e^{i \vec{k} \cdot (\vec{R}_i - \vec{R}_B ) } t_{Bi}  = E C_B (\vec{k})  $$

$$ C_B (\vec{k}) e^{i \vec{k} \cdot (-acc) \hat{i} } t + C_A (\vec{k}) e^{i \vec{k} \cdot \left(3 \frac{acc}{2} \hat{i} + \frac{\sqrt{3}}{2} acc \hat{j} - acc \hat{i} \right)} t + C_A (\vec{k}) e^{i \vec{k} \cdot \left(3\frac{acc}{2} \hat {i} - \frac{\sqrt{3}}{2} acc \hat{j} - acc \hat{i} \right)} t + C_B (\vec{k}) E_s = E C_B (\vec{k}) $$

$$ C_B (\vec{k}) \left( 
e^{i \vec{k} \cdot (-acc) \hat{i} } + 
e^{i \vec{k} \cdot \left(\frac{acc}{2} \hat{i} + \frac{\sqrt{3}}{2} acc \hat{j} \right)} + 
e^{i \vec{k} \cdot \left(\frac{acc}{2} \hat{i} - \frac{\sqrt{3}}{2} acc \hat{j} \right)} 
\right) t 
+ C_B (\vec{k}) E_s = E C_B (\vec{k}) $$

$$ \left( \begin{matrix} Es & AB \\ BA & E_s \end{matrix} \right) 
\left( \begin{matrix} C_A \\ C_B \end{matrix} \right) = E \left( \begin{matrix} C_A \\ C_B \end{matrix} \right) $$


```python
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

acc=1.44
Es = 0.0
t = -2.8

a1 = [1.5*acc, np.sqrt(3)/2*acc, 0.0]
a2 = [1.5*acc, -np.sqrt(3)/2*acc, 0.0]
a3 = [0., 0., 10.]

d = abs(np.dot(a1, np.cross(a2,a3)))
b1 = 2*np.pi*np.cross(a2,a3)/d
b2 = 2*np.pi*np.cross(a3,a1)/d

x = np.linspace(-1,1,50)
y = np.linspace(-1,1,50)

E1 = np.zeros((len(x),len(y)))
E2 = np.zeros((len(x),len(y)))

for i in range(len(x)):
    for j in range(len(x)):
        k = x[i]*b1+y[j]*b2

        AB = (  (np.exp(1j*np.dot(k, [-acc,0,0]))
              + np.exp(1j*np.dot(k, [acc/2, np.sqrt(3)/2*acc,0]))
              + np.exp(1j*np.dot(k, [acc/2,-np.sqrt(3)/2*acc,0]))) *t)

        BA = ( ( np.exp(1j*np.dot(k, [acc,0,0]))
              + np.exp(1j*np.dot(k, [-acc/2,np.sqrt(3)/2*acc,0]))
              + np.exp(1j*np.dot(k, [-acc/2,-np.sqrt(3)/2*acc,0])))*t)

        A = [[Es, AB],
             [BA, Es]]

        ener, eig = np.linalg.eigh(A)
        E1[i][j] = ener[0]
        E2[i][j] = ener[1]
```

```python
%matplotlib inline
fig = plt.figure(dpi=150)
ax = fig.gca(projection='3d')
ax.set_aspect('equal', 'box')

X, Y = np.meshgrid(x, y)
surf = ax.plot_surface(X, Y, E1, cmap=cm.Blues_r,
                       linewidth=0, antialiased=False)
surf = ax.plot_surface(X, Y, E2, cmap=cm.Oranges,
                       linewidth=0, antialiased=False)

ax.set_xticklabels(ax.get_xticks(), fontsize=7.5)
ax.set_yticklabels(ax.get_yticks(), fontsize=7.5)
ax.set_zticklabels(ax.get_zticks(), fontsize=7.5)

ax.set_title('Bandas de energía del grafeno', fontsize=8)

plt.show()
```
